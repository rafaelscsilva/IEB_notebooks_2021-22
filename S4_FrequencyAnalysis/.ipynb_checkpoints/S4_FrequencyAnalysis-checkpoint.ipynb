````python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
````

````python
def plot_spectrum(y, fs):
    N = len(y)                       
    k = np.arange(N)
    T = N/fs
    frq = k/T # two sides frequency range
    freq = frq[range(int(N/2))]           # one side frequency range

    Y = np.fft.fft(y)/N              # fft computing and normalization
    Y = Y[range(int(N/2))]
    
    plt.figure()
    plt.plot(freq, abs(Y), 'r')
    plt.xlabel('Freq (Hz)')
    plt.ylabel('Amplitude')
````

tentar usar exemplos de filtros em harmonia com a documentaçao (bandpass e notch)
filtrar 50hz e 60 hz

````python
fs = 1000                       # sampling rate
Ts = 1/fs                     # sampling interval
T = 10                          # signal duration
t = np.arange(0,T,Ts) 

f1 = 50                          
y1 = np.sin(2 * np.pi * f1 * t)

plot_spectrum(y1, fs)
````

````python
f2 = 25
y2 = np.sin(2 * np.pi * f2 * t)

plot_spectrum(y2, fs)
````

````python
y3 = np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t)

plt.plot(y3[0:100])
plot_spectrum(y3, fs)
````
